# Circular Linked List

```python
class Node:
    def __init__(self,data,next =None):
        self.data = data
        self.next = None
class Circular:
    def get_last(self,to_print = False):
        node = self.head
        while(node.next != self.head):
            if to_print:
                print(node.data)
            node = node.next
        return node
    def __init__(self,head):
        self.head = head
        self.count = 0
        self.head.next = self.head
    def add_node(self,new_node):
        node = self.get_last()
        node.next = new_node
        new_node.next  = self.head
    def print_lk_lst(self):
        node = self.get_last(to_print=True)
        print(node.data)
    def del_last_ele(self):
        node = self.head
        prev = None
        while(node.next != self.head):
            prev = node
            node = node.next
        prev.next = self.head

node1 = Node("Monday")
dl1 = Circular(node1)
dl1.add_node(Node("Tuesdya"))
dl1.add_node(Node("Wednesday"))
dl1.print_lk_lst()
dl1.del_last_ele()
dl1.print_lk_lst()
    
```

# Unrolled LinkList

```python
class SubNode:
    def __init__(self,data):
        self.data = data
        self.next = None
class subUnrolled:
    def __init__(self,head):
        self.head = head
    def sub_get_last(self,to_print = False):
        node = self.head
        while(node.next != None):
            if to_print:
                print(node.data , end=",")
            node = node.next
        return node
    def add_node(self,new_node):
        node = self.sub_get_last()
        node.next = new_node
    def print_subUll(self):
        node = self.sub_get_last(to_print= True)
        print (node.data)
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.subUnrld = subUnrolled(SubNode(data))
class Unrolled:
    def __init__(self,head):
        self.head = head
    def get_last(self,to_print = False):
        node = self.head
        while(node.next != None):
            if to_print:
                print(node.data,end=" Sub-List ==>")
                node.subUnrld.print_subUll()
            node = node.next
        return node
    def add_node(self,new_node):
        node = self.get_last()
        node.next = new_node
    def add_sub_node(self,parent_li,new_node):
        node = self.head
        i = 0
        while(i != parent_li):
            i+=1
            node = node.next
        node.subUnrld.add_node(SubNode(new_node))
    def print_ul(self):
        node = self.get_last(True)
        print(node.data , end=" Sub-List ==>")
        node.subUnrld.print_subUll()
ulstart = Unrolled(Node(10))
ulstart.add_sub_node(0,11)
ulstart.add_sub_node(0,12)
ulstart.add_sub_node(0,13)
ulstart.add_node(Node(20))
ulstart.add_sub_node(1,21)
ulstart.print_ul()
```

# XOR Linked List

An ordinary Doubly Linked List requires space for two address fields to store the addresses of previous and next nodes. A memory efficient version of Doubly Linked List can be created using only one space for address field with every node. This memory efficient Doubly Linked List is called XOR Linked List or Memory Efficient as the list uses bitwise XOR operation to save space for one address. In the XOR linked list, instead of storing actual memory addresses, every node stores the XOR of addresses of previous and next nodes.

Ordinary Representation:
Node A:
prev = NULL, next = add(B) // previous is NULL and next is address of B

Node B:
prev = add(A), next = add(C) // previous is address of A and next is address of C

XOR List Representation:
Let us call the address variable in XOR representation npx (XOR of next and previous)

Node A:
npx = 0 XOR add(B) // bitwise XOR of zero and address of B

Node B:
npx = add(A) XOR add(C) // bitwise XOR of address of A and address of C

Node C:
npx = add(B) XOR add(D) // bitwise XOR of address of B and address of D
