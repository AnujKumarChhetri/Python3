# Bubble Sort

```python
from IPython.core.display import Image, display
display(Image('bubble_sort.png', width=290, unconfined=True))
```

```python
#default_method
# Using 2 for loop
# Complexity--> worst(O(n^2))
# best --> (O(n))
```

```python
#my_method
def bubble_sort(li):
    swap = False
    for x in range(len(li)-1):
        if li[x] > li[x+1]:
            swap = True
            temp = li[x+1]
            li[x+1] = li[x]
            li[x] = temp
    if not swap:
        return li
    else:
        return bubble_sort(li)
print(bubble_sort([1,2,3,1,45,23,5,2,3,5,1,3,45,5,6,7]))
# complexity --> (O(n))
```

Complexity of bubble sort is o(n^2) but we can compress to o(n) <br>
Recursive bubble sort is a comparison based sort with Complexity -- > O(n)

# Selection Sort 

```python
from IPython.core.display import Image, display
display(Image('selection_Sort.png', width=290, unconfined=True))
```

```python
def swaping(A,x,y):
    temp = A[y]
    A[y] = A[x]
    A[x] = temp
```

```python
#default method
def selection_sort(li):
    for x in range(len(li)):
        small = x
        for y in range(small+1, len(li)):
            if li[y] < li[small]:
                small = y
        swaping(li,x,small)
    return li
print(selection_sort([1,2,31,4,1,41,5]))
```

Complexity --> O(n^2) For both , best and worst

# Insertion Sort

```python
from IPython.core.display import Image, display
display(Image('insertionsort.png', width=290, unconfined=True))
```

```python
#Implementaion done for first loop of increment and second loop of decrement
#Reason for second loop to be in decrement order is 
#--> Check the case of 10
#--> if we check 10 from 2,3,4,5 its like o(n)
#--> but if we check 10 from back with 4 we dont need to traverse the whole array again
```

```python
def insertion_sort(li):
    for x in range(1,len(li)):
        temp = li[x]
        y = x
        
#       while li[y-1] > temp and y != 0:
#           swaping(li,y-1,y) 
#           y -=1
#       OR WE CAN ALSO DO THIS LIKE

        while li[y-1] > temp and y!=0:
            li[y] = li[y-1]
            y-=1
        li[y] = temp
    return li
print(insertion_sort([5,4,3,2,1]))
```

# Merge Sort

```python
from IPython.core.display import Image, display
display(Image('mergrsort.png', width=290, unconfined=True))
```

ALGO for my own <br>
1 - Divide left until you dont get len(1)  and store them for sorting<br>
2 - divide right until you dont get len(1) and sotore them for sorting<br>
3 - When get none in return from right in end case for ex let say <br>
    i --> call rigth break [9,2] it will call as [9] only  <br>
    ii --> Then statement false and return none so the list is [9,2] and follow the sorting  <br>
    iii --> return the sorted(in case of return method) array which will then return to the rigth call which id done when li is [9,82,10]
Now this all will called recursively <br>
HEre sorting is done basicly in a way like merge 2 sorted array, in sorted form

```python
def merge_sort(li):
    if len(li) >1:
        mid = round(len(li)/2)
        right_li = li[mid:]
        left_li = li[:mid]
        #--> step 1 left break means its right list
        merge_sort(right_li)
        #--> step 2 right break means its left list
        merge_sort(left_li)
        #i for left, j for right, k for position to insert
        i=j=k = 0
        while i<(len(left_li)) and j<(len(right_li)):
            if left_li[i]<right_li[j]:
                li[k] = left_li[i]
                k+=1
                i+=1
            else:
                li[k] = right_li[j]
                j+=1
                k+=1
        while i< len(left_li):
            li[k] = left_li[i]
            k+=1
            i+=1
        while j< len(right_li):
            li[k] = right_li[j]
            k+=1
            j+=1
#         return li
    else:
        return None
```

# Quick Sort

```python
from IPython.core.display import Image, display
display(Image('quicksort.png', width=390, unconfined=True))
```

```python
def break_and_sort(li,low,high):
    pivot = high
    for x in range(low,high):
        if li[x] <= li[pivot]:
            li[x],li[low] = li[low],li[x]
            low+=1
    li[low],li[pivot] = li[pivot],li[low]
    return low
def quicksort(li,start = 0,end = None):
    if end == None:
        end = len(li)-1
    if start<end:
        pivot = break_and_sort(li,start,end)
        #for right array
        quicksort(li,pivot+1,end)
        #for right array
        quicksort(li,0,pivot-1)
```

```python
a = [9,-3,5,2,6,8,-6,3]
# a = [-3, 2, -6, 3, 6, 8, 5, 9]
quicksort(a)
print(a)
```
