# Stack And Queue

```python
def print_block(data):
    print(" ↑ ")
    print(' _ ')
    print(f"|{data}|")
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
```

```python
class Stack:
    def __init__(self,first_node,max_items):
        self.head = Node(first_node)
        self.max_items = max_items
        self.len = 1
    #Stack folows last in first out (LIFO)
    #means in stack element add to head and pop from head so to maintain complexity of o(1)
    def push(self,node):
        if self.head == None:
            self.head = Node(node)
            self.len +=1
        elif self.len < self.max_items:
            current = self.head
            self.head = Node(node)
            self.head.next = current
            self.len+=1
        else:
            raise Exception("Stack Memory is Full")
    def pop(self):
        if self.len == 0:
            raise Exception("Stack Is emplty")
        else:
            self.len -=1
            deleted_node = self.head
            self.head = self.head.next
            current = self.head
        while current.next != None:
            if current == self.head:
                print("Stack   POped Item")
                print(' _ ')
                print(f"|{self.head.data}|  -->  |{deleted_node.data}|")
            else:
                print_block(current.data)
            current = current.next
        print_block(current.data)
    def print_stack(self):
        current = self.head
        while current.next != None:
            if current == self.head:
                print("Stack")
                print(' _ ')
                print(f"|{current.data}|")
            else:
                print_block(current.data)
            current = current.next
        print_block(current.data)
    def __len__(self):
        return self.len
stack = Stack(1,6)
for x in range(2,7):
    try:
        stack.print_stack()
        stack.push(x)
        print()
    except Exception as e:
        print(e)
        break
stack.print_stack()
print()
stack.pop()
```

```python
class Queue:
    def __init__(self,first_node,max_items):
        self.head = Node(first_node)
        self.max_items = max_items
        self.tail = None
        self.len = 1
    #Queue folows First in first out (LIFO)
    #means we can aslo use tail for the last element in link list so we deque element with 0(1)
    def enqueue(self,node):
        if self.head == None:
            node = Node(node)
            self.head = node
            self.tail = node
            self.len +=1
        elif self.len < self.max_items:
            node = Node(node)
            if self.tail == None:
                self.head.next = node
                self.tail  = node
            else:
                temp = self.tail
                self.tail.next = node
                self.tail = self.tail.next
            self.len +=1
        else:
            raise Exception("Queue Memory is Full")
    def dequeue(self):
        if self.len == 0:
            raise Exception("Queue Is emplty")
        else:
            deleted_node = self.head
            self.head = self.head.next
            current = self.head
            self.len -=1
        while current.next != None:
            if current == self.head:
                print("Queue   POped Item")
                print(' _ ')
                print(f"|{self.head.data}|  -->  |{deleted_node.data}|")
            else:
                print_block(current.data)
            current = current.next
        print_block(current.data)
    def print_queue(self):
        current = self.head
        while current.next != None:
            if current == self.head:
                print("Queue")
                print(' _ ')
                print(f"|{current.data}|")
            else:
                print_block(current.data)
            current = current.next
        print_block(current.data)
    def __len__(self):
        return self.len
queue = Queue(1,6)
for x in range(2,7):
    try:
        queue.print_queue()
        print()
        queue.enqueue(x)
    except Exception as e:
        print(e)
        break
queue.print_queue()
print()
queue.dequeue()
```
