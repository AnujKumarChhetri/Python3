# 3 index swaping

```python
def swap(li,x,y,z):
    a = li[x]
    li[x] = li[y]
    li[y] = li[z]
    li[z] = a
    return(li)

def three_sort(li):
    i = 0
    len_li = len(li)
    while(len_li>=3):
        x = i
        y = i+1
        z = i+2
        if li[y]>li[z]<li[x]>=li[y]:
            print('a')
            swap(li,x,y,z)
            swap(li,x,y,z)
            i+=1
            len_li-=1
        elif li[z]<li[x]<li[y]:
            print('b')
            li = swap(li,x,y,z)
            li = swap(li,x,y,z)
            i+=1
            len_li -=1
        elif li[z]<=li[x]>li[y]:
            print('c')
            li = swap(li,x,y,z)
            i+=1
            len_li -=1
        elif li[x]<=li[y]:
            print('d')
            i+=1
            len_li -=1
        else:
            print(li,'Errro')
            break    
        return(li)
if __init__ == main():
        li = list(map(int,input().split(',')))
        three_sort(li)
```
